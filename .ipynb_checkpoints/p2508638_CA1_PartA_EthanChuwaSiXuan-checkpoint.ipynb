## **1. Problem Definition & Task**

<hr>

This project builds a machine learning model to predict whether a factory machine is in a **normal** or **abnormal** condition based on sensor readings and machine information.

* **Task type:** *Binary Classification* (0 = normal, 1 = abnormal)
* **Goal:** Build a model that accurately detects abnormal machine conditions and generalises well to unseen data.
* **Target variable:** **`Machine Status`**

  * **0** → Normal
  * **1** → Abnormal
 **Key performance metrics**

- **Recall (Class 1 – main metric)**  
  Measures how many faulty machines are successfully detected  
  *(TP / (TP + FN))*. This is the highest priority for safety and maintenance.

- **F1-score (weighted)**  
  Balances precision and recall but is still dominated by the majority class.

- **Precision**  
  Measures how many predicted faults are actually real faults  
  (helps avoid false alarms).

- **Accuracy**  
  Shows overall correctness but becomes **misleading on imbalanced datasets**.

- **ROC-AUC**  
  Measures how well the model separates faulty vs. normal machines across thresholds.

<hr>

 **Why Recall is the main metric**

In predictive maintenance, **missing a faulty machine (false negative)** is much worse than raising a false alarm and can lead to:

- machine failure  
- unplanned downtime  
- high repair costs  
- potential safety risks  

Because of this, **minimizing false negatives is more important than minimizing false positives**.  
Recall directly measures how many faults the model correctly identifies, making it the most meaningful metric for evaluating model performance in this context.

**Therefore, Recall₁ (fault detection recall) is the primary evaluation metric for this project.**

##  **3. Data Cleaning, Preprocessing, Train Test Split**
<hr>
This section will focus on preparing data.

### **3.1 Dropping unusable data**
Since there are NaN values in Quality, I will drop their rows as the data does not contribute in any meaningful way.

```python
df.isnull().sum().sort_values(ascending=True)
```

```python
df = df.dropna(subset=["Quality"])
df.isnull().sum().sort_values(ascending=True)
```

### **3.2 Train-Test Split**

```python
X=df[num_cols]
y=df["Machine Status"]
X_train, X_test, y_train, y_test = train_test_split(
    X,y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)
```

`df` was split into `X_train`,`X_test`,`y_train`,`y_test`.

### **3.3 Imputation, UnderSampling and Scaling**
* Since there are NaN values, imputation is carried out.
* Undersampling is carried out to solve imbalanced dataset.
* Scaling is performed to standardize the numerical features.

`SimpleImputer()` is then used the impute median into NaN values inside `X_train` and `X_test`.

```python
imputer = SimpleImputer(strategy="median")

X_train=imputer.fit_transform(X_train)
X_test=imputer.transform(X_test)
X_train = pd.DataFrame(X_train, columns=num_cols, index=y_train.index)
X_test = pd.DataFrame(X_test, columns=num_cols, index=y_test.index)
```

Median imputation is used because the median is robust to outliers and preserves the overall data distribution. Replacing missing values with the median introduces minimal distortion to the dataset and ensures that the central tendency of each feature remains consistent.

We then use `isnull()` to confirm that there are no more NaN values left.

```python
print(X_test.isnull().sum().sort_values(ascending=True))
print(X_train.isnull().sum().sort_values(ascending=True))
```

We can see that there are no NaN values left.

Solving imbalance

* The dataset is **highly imbalanced** (many normal machines, very few faulty machines).
* A model trained on this imbalance tends to **always predict class 0**, resulting in high accuracy but **extremely poor recall** for detecting faults.
* Undersampling reduces the number of normal samples so both classes are **balanced during training**, forcing the model to learn patterns from the minority (faulty) class.
* This improves the model’s ability to detect faults by **reducing bias toward the majority class**.
* Oversampling was **not chosen** because duplicating minority samples increases the risk of **overfitting**, especially given the small number of fault cases.
* Undersampling provides a **cleaner, simpler, and safer** way to improve minority-class recall without artificially inflating the dataset.


```python
train_df = pd.concat([X_train, y_train], axis=1) 
under_minority = train_df[train_df['Machine Status'] == 1] 
under_majority = train_df[train_df['Machine Status'] == 0].sample( 
    n=len(under_minority),
    random_state=RANDOM_STATE
)
undersampled_df = pd.concat([under_minority, under_majority]).sample(frac=1,random_state=RANDOM_STATE)
X_train_undersampled = undersampled_df[['Ambient T (C)','Process T (C)','Rotation Speed (rpm)','Torque (Nm)','Tool Wear (min)','Quality']]
y_train_undersampled = undersampled_df[['Machine Status']]
```

* Random undersampling reduces the majority class to match the minority class.
* This creates a balanced training set where class 0 and class 1 have equal samples.
* The model now learns patterns for both classes, improving fault detection (Recall₁), which is the main metric of this project.

Features were on very different scales (e.g., Rotation Speed in thousands, Temperature in tens). This causes distance-based and gradient-based models to be biased toward larger-scaled features. Hence `StandardScaler()` was used to scale the data.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_undersampled_scaled = scaler.transform(X_train_undersampled)

X_train_scaled_df = pd.DataFrame(
    X_train_scaled, columns=X_train.columns, index=X_train.index
)

X_test_scaled_df = pd.DataFrame(
    X_test_scaled, columns=X_test.columns, index=X_test.index
)

X_train_undersampled_scaled_df = pd.DataFrame(
    X_train_undersampled_scaled, 
    columns=X_train.columns, 
    index=X_train_undersampled.index
)
```

All features were standardized using `StandardScaler`, converting them to mean = 0 and std = 1. This ensures equal contribution of all features, prevents scale dominance, improves model performance, and applies the same scaling consistently to train, test, and undersampled datasets.

## **4. Baseline Model & Model Training & Evaluation**
<hr>

### **4.1 Setting up KFold**

I use `StratifiedKFold` to create a 5-fold cross-validation splitter that randomly shuffles the data before dividing it into folds to ensure fair and reliable model evaluation.

```python
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
```

### **4.2 Dummy Classifier**

A `DummyClassifier(strategy="most_frequent")` is evaluated using cross-validation across accuracy, precision, recall, F1, and ROC-AUC.
This baseline always predicts the majority class (0 = normal), allowing us to measure how a trivial model performs.

```python
scoring = [
    "accuracy",
    "precision",
    "recall",
    "f1",
    "roc_auc"
]
baseline_results = {}
baseline = DummyClassifier(strategy="most_frequent", random_state=RANDOM_STATE)

for metric in scoring:
    scores = cross_val_score(
        baseline,
        X_train_scaled,
        y_train,
        cv=kf,
        scoring=metric
    )
    mean_score = scores.mean()
    std_score = scores.std()

    # Store results
    baseline_results[metric] = {
        "scores": scores,
        "mean": mean_score,
        "std": std_score
    }
    
    print(f"Baseline {metric} scores (CV): {scores}")
    print(f"Baseline mean {metric}: {scores.mean():.4f} (+/- {scores.std():.4f})\n")
```

The DummyClassifier baseline, which always predicts the majority class, achieves a high accuracy of 0.9661, but its precision, recall, and F1-score are all 0.0 and ROC-AUC is 0.5. This happens because the model never predicts any machines as faulty (class 1), leading to no true positives and all actual faults being missed. The undefined precision warnings confirm this behaviour. These results highlight that accuracy is misleading for this imbalanced problem.

### **4.3 Model Evaluation**

Next, we compare four classifiers (Logistic Regression, KNN, Decision Tree, Random Forest) using cross-validation on both the original imbalanced dataset and the balanced undersampled dataset. For each model and each evaluation metric, the code computes cross-validated scores and visualizes how class imbalance affects model behaviour. This helps identify which model improves the most when class imbalance is corrected, and which model is most promising for detecting faulty machines.

```python
models = {
    "Logistic Regression": LogisticRegression(random_state=RANDOM_STATE, max_iter=1000),
    "KNN Classifier": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=RANDOM_STATE),
    "Random Forest": RandomForestClassifier(random_state=RANDOM_STATE)
}

y_train_under_1d = y_train_undersampled.values.ravel()

for model_name, model in models.items():
    print("\n======================================")
    print(f"Model: {model_name}")
    print("======================================")
    orig_means = []
    under_means = []
    for metric in scoring:
        scores_orig = cross_val_score(
            model,
            X_train_scaled,
            y_train,
            cv=kf,
            scoring=metric,
            n_jobs=-1
        )

        scores_under = cross_val_score(
            model,
            X_train_undersampled_scaled,
            y_train_under_1d,
            cv=kf,
            scoring=metric,
            n_jobs=-1
        )

        print(f"\nMetric: {metric}")
        print(f"Original      {metric} scores (CV): {scores_orig}")
        print(f"Original      mean {metric}: {scores_orig.mean():.4f} (+/- {scores_orig.std():.4f})")

        print(f"Undersampled  {metric} scores (CV): {scores_under}")
        print(f"Undersampled  mean {metric}: {scores_under.mean():.4f} (+/- {scores_under.std():.4f})")
        orig_means.append(scores_orig.mean())
        under_means.append(scores_under.mean())
    x = np.arange(len(scoring))      
    width = 0.35                      

    plt.figure(figsize=(10, 5))
    plt.bar(x - width/2, orig_means, width, label="Original Train")
    plt.bar(x + width/2, under_means, width, label="Undersampled Train")

    plt.xticks(x, scoring)
    plt.ylabel("CV Score")
    plt.title(f"Cross-Validation Comparison for {model_name}")
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)
    plt.tight_layout()
    plt.show()
```


 **Summary of Model Comparison**

* Before undersampling, **Logistic Regression (0.186)** and **KNN (0.399)** performed poorly on Recall₁, meaning they failed to detect most faulty machines due to the severe class imbalance.
* Tree-based models handled imbalance better, but still missed faults: **Decision Tree (0.816)** and **Random Forest (0.764)** were stronger but not sufficient for a safety-critical system.
* After undersampling, every model showed a **major improvement in Recall₁**, demonstrating that balancing the data allows the algorithms to correctly learn fault patterns.

  * **Logistic Regression:** 0.186 → **0.800**
  * **KNN:** 0.399 → **0.888**
  * **Decision Tree:** 0.816 → **0.907**
  * **Random Forest:** 0.764 → **0.957**
* These improvements directly reduce **false negatives**, which is the most important goal in predictive maintenance.
* Among all models, **Random Forest achieved the highest Recall₁ (0.957)** while maintaining strong F1 and ROC-AUC, making it the most reliable and safest model for detecting abnormal machine conditions.
* Hence we will continue our evaluation using only the undersampled data set as it is shown to consistently improve recall over all models.



Since Recall is our primary metric, lets plot a bar graph to compare recall scores across the different models better.

```python
model_names = []
recall_means = []

for model_name, model in models.items():
    scores = cross_val_score(
        model,
        X_train_undersampled_scaled,
        y_train_under_1d,
        cv=kf,
        scoring="recall",
        n_jobs=-1
    )
    model_names.append(model_name)
    recall_means.append(scores.mean())


plt.figure(figsize=(8, 5))
bars = plt.bar(model_names, recall_means)


for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 0.01,                 
        f"{height:.3f}",               
        ha='center', va='bottom'
    )

plt.ylabel("Mean Recall (CV)")
plt.title("Recall Comparison Across Models (Undersampled Data)")
plt.xticks(rotation=20)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

Random Forest achieving the highest recall (0.957), meaning it correctly identifies almost all faulty machines. Decision Tree follows closely at 0.907, showing strong fault-detection ability as well. KNN performs moderately well with a recall of 0.888, while Logistic Regression, although improved from the imbalanced case, remains the weakest at 0.800. This visual confirms that tree-based models—especially Random Forest—learn fault patterns most effectively when trained on balanced data.

This following then evaluates all four models using cross-validated predictions on the undersampled training set. It produces a confusion matrix for each model and stores ROC curve data (FPR, TPR, AUC). After looping through all models, it generates one final combined ROC plot showing the comparative performance of each classifier under the balanced training scenario.

```python
roc_results = [] 


for model_name, model in models.items():
    print("\n" + "="*50)
    print(f"{model_name} – Undersampled (CV predictions)")
    print("="*50)

  
    y_pred_cv = cross_val_predict(
        model,
        X_train_undersampled_scaled,
        y_train_under_1d,
        cv=kf,
        n_jobs=-1
    )

    y_proba_cv = cross_val_predict(
        model,
        X_train_undersampled_scaled,
        y_train_under_1d,
        cv=kf,
        method="predict_proba",
        n_jobs=-1
    )[:, 1]


    cm = confusion_matrix(y_train_under_1d, y_pred_cv)
    print("Confusion Matrix:\n", cm)

    disp = ConfusionMatrixDisplay(cm)
    disp.plot(cmap="Greens")
    plt.title(f"Confusion Matrix – {model_name} (Undersampled, CV)")
    plt.tight_layout()
    plt.show()


    fpr, tpr, _ = roc_curve(y_train_under_1d, y_proba_cv)
    auc = roc_auc_score(y_train_under_1d, y_proba_cv)
    roc_results.append((model_name, fpr, tpr, auc))

plt.figure(figsize=(10, 7))

for model_name, fpr, tpr, auc in roc_results:
    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {auc:.3f})")

plt.plot([0, 1], [0, 1], "k--", label="Random Guess")

plt.title("ROC Curves for All Models (Undersampled, CV Predictions)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The confusion matrices and combined ROC plot show clear differences in robustness. Logistic Regression still struggles, missing over 100 faulty machines, while KNN significantly reduces false negatives. Decision Tree performs even better, showing a strong balance with low FP/FN values. Random Forest, however, stands out as the best-performing model: it achieves the fewest false negatives (22) and the highest ROC-AUC (0.973), demonstrating superior ability to separate normal and faulty machine conditions. This makes Random Forest the most effective and reliable model for predictive maintenance, where maximizing Recall₁ and minimizing false negatives is critical.

We will then proceed with hyperparameter tuning for Decision Tree and Random Forest.

## **5. Hyperparameter Tuning**
<hr>

### **5.1 Hyperparameter tuning for Decision Tree and Random Forest**

The following performs hyperparameter tuning for both Decision Tree and Random Forest using GridSearchCV as they are the best 2 performing models from model evaluation previously. A grid of parameters is defined for each model, and the tuning process is driven by the recall metric to prioritise detecting faulty machines. GridSearchCV uses cross-validation on the undersampled training data to ensure fair, balanced learning, and it evaluates every parameter combination to find the settings that maximise Recall₁. The output of this step is the best-tuned version of each model, along with their highest cross-validated recall scores, which will later be compared to identify the most effective model for predictive maintenance.

```python
dt = DecisionTreeClassifier(random_state=RANDOM_STATE)

param_grid_dt = {
    "criterion": ["gini", "entropy"],
    "max_depth": [None, 5, 10, 15],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

grid_dt = GridSearchCV(
    estimator=dt,
    param_grid=param_grid_dt,
    scoring="recall",      
    cv=kf,
    n_jobs=-1,
    verbose=1
)

grid_dt.fit(X_train_undersampled_scaled, y_train_under_1d)

best_dt = grid_dt.best_estimator_
print("\n===== Best Decision Tree =====")
print("Best Params:", grid_dt.best_params_)
print("Best CV Recall:", grid_dt.best_score_)



rf = RandomForestClassifier(random_state=RANDOM_STATE)

param_grid_rf = {
    "n_estimators": [100, 200, 300],
    "max_depth": [None, 5, 10, 15],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2", None]
}

grid_rf = GridSearchCV(
    estimator=rf,
    param_grid=param_grid_rf,
    scoring="recall",
    cv=kf,
    n_jobs=-1,
    verbose=1
)

grid_rf.fit(X_train_undersampled_scaled, y_train_under_1d)

best_rf = grid_rf.best_estimator_
print("\n===== Best Random Forest =====")
print("Best Params:", grid_rf.best_params_)
print("Best CV Recall:", grid_rf.best_score_)
```

The hyperparameter tuning results show that both tree-based models improved further when optimized specifically for recall on the balanced training data. The Decision Tree performed best with entropy, a shallow depth of 5, and slightly larger leaf size settings, achieving a strong cross-validated recall of 0.9613. Meanwhile, the Random Forest achieved the highest overall performance, with its best configuration using 200 trees, a max_depth of 5, and small leaf sizes, resulting in the best CV recall of 0.9632. These results confirm that controlled tree depth and slightly more conservative splits help the models generalize better while still detecting nearly all faulty machines. Overall, the tuned Random Forest remains the strongest model, showing the highest recall and reinforcing its suitability for fault-detection tasks where minimizing false negatives is critical.

### **5.2 Comparing tuned vs untuned results**


The following then compares how much **hyperparameter tuning** improves the **recall** of the Decision Tree and Random Forest models on the undersampled (balanced) training data. First, it defines **untuned** versions of both models and uses `cross_val_score` with `scoring="recall"` to compute their mean cross-validated recall scores. Next, it takes the **tuned models** (`best_dt` and `best_rf`) obtained earlier from `GridSearchCV` and evaluates them in exactly the same way, again using cross-validated recall on the same undersampled data. The resulting four recall values—untuned DT, tuned DT, untuned RF, tuned RF—are then visualised in a grouped bar chart, where each pair of bars shows how recall changes before vs after tuning for each model. This plot makes it easy to see whether tuning actually improves fault-detection performance and which model benefits the most from hyperparameter optimisation.


```python


untuned_dt = DecisionTreeClassifier(random_state=RANDOM_STATE)
untuned_rf = RandomForestClassifier(random_state=RANDOM_STATE)


untuned_dt_scores = cross_val_score(
    untuned_dt,
    X_train_undersampled_scaled,
    y_train_under_1d,
    cv=kf,
    scoring="recall",
    n_jobs=-1
)
untuned_dt_recall = untuned_dt_scores.mean()


untuned_rf_scores = cross_val_score(
    untuned_rf,
    X_train_undersampled_scaled,
    y_train_under_1d,
    cv=kf,
    scoring="recall",
    n_jobs=-1
)
untuned_rf_recall = untuned_rf_scores.mean()



best_dt = grid_dt.best_estimator_
best_rf = grid_rf.best_estimator_


tuned_dt_scores = cross_val_score(
    best_dt,
    X_train_undersampled_scaled,
    y_train_under_1d,
    cv=kf,
    scoring="recall",
    n_jobs=-1
)
tuned_dt_recall = tuned_dt_scores.mean()


tuned_rf_scores = cross_val_score(
    best_rf,
    X_train_undersampled_scaled,
    y_train_under_1d,
    cv=kf,
    scoring="recall",
    n_jobs=-1
)
tuned_rf_recall = tuned_rf_scores.mean()




labels = ["Decision Tree", "Random Forest"]
untuned_values = [untuned_dt_recall, untuned_rf_recall]
tuned_values = [tuned_dt_recall, tuned_rf_recall]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(9, 6))

bars1 = plt.bar(x - width/2, untuned_values, width, label="Untuned", color="skyblue")
bars2 = plt.bar(x + width/2, tuned_values, width, label="Tuned", color="orange")


for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + 0.01,
            f"{height:.3f}",
            ha="center",
            va="bottom"
        )

plt.xticks(x, labels)
plt.ylabel("Mean Recall (CV)")
plt.title("Recall Comparison: Tuned vs Untuned Models")
plt.ylim(0, 1)
plt.legend()
plt.tight_layout()
plt.show()
```



The tuned models show clear improvements in recall compared to their untuned versions, demonstrating that hyperparameter optimisation enhances their ability to detect faulty machines. The key observations are:

**Decision Tree**

* **Untuned Recall:** 0.907
* **Tuned Recall:** 0.961
* **Improvement:** Significant
* Tuning helped the tree create more meaningful splits and reduce misclassified faults.

**Random Forest**

* **Untuned Recall:** 0.957
* **Tuned Recall:** 0.963
* **Improvement:** Small but consistent
* Random Forest was already strong before tuning, so gains are incremental but still positive.

**Overall Insight**

* Hyperparameter tuning benefits both models, but **Decision Tree gains more** because it initially had more room for optimisation.
* **Random Forest remains the top-performing model** with the highest recall after tuning.
* This reinforces Random Forest as the **best and safest choice for predictive maintenance**, where **minimising false negatives** is the most important goal.

### **6.2 Comparison of CV and test results**



The following compares the tuned Random Forest model’s performance during training and on unseen data by:

* **Defining a consistent set of metrics** (accuracy, precision, recall, F1, ROC-AUC) to ensure both evaluations use the same criteria.
* **Computing cross-validation scores** on the undersampled training set, which shows how well the tuned model performs under balanced conditions and whether it generalises across folds.
* **Evaluating the same tuned model on the test set**, generating predictions and calculating all metrics on real unseen data.
* **Building a side-by-side comparison table** that places cross-validation results next to test-set results, making it easy to assess generalisation, consistency, and potential overfitting.
* This comparison confirms whether the tuned Random Forest maintains strong recall and overall performance beyond training, ensuring it is reliable for real predictive maintenance use.


```python

metrics = ["accuracy", "precision", "recall", "f1", "roc_auc"]


cv_means = {}

for metric in metrics:
    scores = cross_val_score(
        best_rf,                      
        X_train_undersampled_scaled,  
        y_train_under_1d,              
        cv=kf,
        scoring=metric,
        n_jobs=-1
    )
    cv_means[metric] = scores.mean()


y_pred = best_rf.predict(X_test_scaled)
y_proba = best_rf.predict_proba(X_test_scaled)[:, 1]

test_results = {
    "accuracy":  accuracy_score(y_test, y_pred),
    "precision": precision_score(y_test, y_pred, zero_division=0),
    "recall":    recall_score(y_test, y_pred, zero_division=0),
    "f1":        f1_score(y_test, y_pred, zero_division=0),
    "roc_auc":   roc_auc_score(y_test, y_proba)
}


comparison_df = pd.DataFrame(
    {
        "CV (Train, undersampled)": cv_means,
        "Test Set": test_results
    }
)

print("\n=== Tuned Random Forest: Cross-Validation vs Test Set ===")
display(comparison_df.T)
```

```python

metrics = ["accuracy", "precision", "recall", "f1", "roc_auc"]

cv_values = [cv_means[m] for m in metrics]
test_values = [test_results[m] for m in metrics]

x = np.arange(len(metrics))
width = 0.35


plt.figure(figsize=(10, 6))

bars1 = plt.bar(x - width/2, cv_values, width, label="CV (Train, Undersampled)",
                color="skyblue")
bars2 = plt.bar(x + width/2, test_values, width, label="Test Set",
                color="orange")


for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + 0.01,
            f"{height:.3f}",
            ha="center", va="bottom"
        )


plt.xticks(x, metrics)
plt.ylabel("Score")
plt.ylim(0, 1)
plt.title("Tuned Random Forest: CV vs Test Set Performance")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)
plt.tight_layout()
plt.show()
```



**After — Tuned Random Forest: CV vs Test Set Performance**

* **Strong generalisation:**

  * Test accuracy (0.903) is close to CV accuracy (0.923), showing the model performs consistently on unseen data and is not overfitting.

* **High fault-detection ability maintained:**

  * Recall remains extremely high on the test set (0.922 vs 0.963 in CV), meaning the model successfully detects most faulty machines even outside the training environment.

* **ROC-AUC remains excellent:**

  * Test ROC-AUC (0.951) is close to CV ROC-AUC (0.969), confirming that the tuned Random Forest maintains strong ability to distinguish normal vs faulty machines across thresholds.

* **Precision and F1 drop on real data:**

  * Precision falls sharply (0.25 vs 0.893 CV), and F1-score decreases (0.393 vs 0.927 CV).
  * This happens because the **test set is highly imbalanced**, unlike the balanced CV folds, leading to more false positives.

* **Trade-off is acceptable for predictive maintenance:**

  * Even with lower precision, the model avoids missing faulty machines (very few false negatives), which is the most important requirement for safety and cost prevention.

* **Final conclusion:**

  * The tuned Random Forest is **stable, generalises well**, and remains highly effective for detecting abnormal machine conditions, making it a reliable final model for deployment.
